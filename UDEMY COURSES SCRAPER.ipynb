```python
from selenium import webdriver
import getpass
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary
from selenium.common import exceptions
import time
from bs4 import BeautifulSoup as bs
import pandas as pd
import random
import csv

url = 'https://www.udemy.com/home/my-courses/learning/'
browser = webdriver.Firefox(executable_path = '') #set path of geckodriver
browser.get(url)

time.sleep(3)
```

```python

names=[]
links=[]


while True:
    next_page=browser.find_elements_by_xpath('/html/body/div[1]/div[3]/div/div/ul/li/div[4]/div/ul/li[8]/a')
    if len(next_page) < 1:
        print("No more pages left")
        break
    else:
        try:
            course_card = browser.find_elements_by_css_selector('div.card.card--learning')
            for course in course_card:
                    course_n = course.find_elements_by_css_selector('strong.details__name')
                    course_l = course.find_elements_by_css_selector('a.card--learning__details')


                    for i in course_n:  
                         for j in course_l:
                            name=i.text
                            href= j.get_attribute('href')
                            names.append(name)
                            links.append(href)


            browser.find_element_by_xpath('/html/body/div[1]/div[3]/div/div/ul/li/div[4]/div/ul/li[8]/a').click()
            time.sleep(10)
        except exceptions.StaleElementReferenceException:
             pass
    print(names)
    print(links)
df = pd.DataFrame(list(zip(names,links)),columns=['Course Name', 'Course Link'])
scraped_data = df.to_csv('udemy_courses.csv', index=False)           
```
